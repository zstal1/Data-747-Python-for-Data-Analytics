1.  Create a line plot of the provided dataset with years on the x-axis and population on the y-axis.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Generate sample data
data = {'Year': [2000, 2005, 2010, 2015, 2020],
        'Population': [6.1, 6.5, 6.9, 7.3, 7.8]}
df = pd.DataFrame(data)

# Create the line plot
plt.plot(df['Year'], df['Population'], marker='o')
plt.title('World Population Growth Over Years')
plt.xlabel('Year')
plt.ylabel('Population (in billions)')
plt.show()
```

5. Create a pie chart of the provided dataset with the percentage of car sales of each brand.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Generate sample data
data = {'Brand': ['Toyota', 'Honda', 'Ford', 'BMW', 'Audi'],
        'Sales': [1500, 1800, 1600, 1400, 1200]}
df = pd.DataFrame(data)

# Create the pie chart
plt.pie(df['Sales'], labels=df['Brand'], autopct='%1.1f%%', startangle=90)
plt.title('Car Sales Percentage by Brand')
plt.axis('equal')
plt.show()
```

6. Create a stacked bar chart of the provided dataset with the years on the x-axis and the total number of medals won by each country on the y-axis. The stacked bars should show the number of gold, silver, and bronze medals won.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Generate sample data
data = {'Year': [2016, 2020],
        'Gold': [100, 110],
        'Silver': [95, 105],
        'Bronze': [110, 120]}
df = pd.DataFrame(data)
df.set_index('Year', inplace=True)

medal_colors = ['gold', 'silver', 'chocolate']

# Create the stacked bar plot
df.plot(kind='bar', stacked=True, color=medal_colors)
plt.title('Total Medals Won in Olympic Years')
plt.xlabel('Year')
plt.ylabel('Total Number of Medals')
plt.xticks(rotation=0)
plt.legend(title='Medal Type')
plt.show()
```

7. Create a box plot of the provided dataset with car prices grouped by brand.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Generate sample data
data = {'Brand': ['Toyota', 'Toyota', 'Toyota', 'Honda', 'Honda', 'Honda', 'BMW', 'BMW', 'BMW'],
        'Price': [24000, 25000, 26000, 21000, 22000, 23000, 44000, 45000, 50000]}
df = pd.DataFrame(data)

# Create the box plot
df.boxplot(column='Price', by='Brand', grid=False)
plt.title('Distribution of Car Price by Brand')
plt.suptitle('')
plt.xlabel('Brand')
plt.ylabel('Price (USD)')
plt.show()
```

8. Create a violin plot of the provided dataset with car prices grouped by brand.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Generate sample data
data = {'Brand': ['Toyota', 'Toyota', 'Toyota', 'Toyota', 'Honda', 'Honda', 'Honda', 'Honda', 'BMW', 'BMW', 'BMW', 'BMW'],
        'Price': [24000, 25000, 26000, 28000, 21000, 22000, 23000, 25000, 44000, 45000, 50000, 52000]}
df = pd.DataFrame(data)

# Prepare data for violin plot
brands = df['Brand'].unique()
price_data = [df['Price'][df['Brand'] == brand] for brand in brands]

# Create the violin plot
fig, ax = plt.subplots()
ax.violinplot(price_data, showmedians=True)
ax.set_title('Distribution of Car Price by Brand')
ax.set_xlabel('Brand')
ax.set_ylabel('Price (USD)')

# Set the x-axis ticks to be the brand names
import numpy as np
ax.set_xticks(np.arange(1, len(brands) + 1))
ax.set_xticklabels(brands)
plt.show()
```

9. Create a hexbin plot of the provided dataset with car prices on the x-axis and mileage on the y-axis.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Generate sample data
data = {'Price': [20000, 22000, 25000, 30000, 35000, 40000, 45000, 50000, 55000, 60000],
        'Mileage': [30, 28, 25, 22, 20, 18, 16, 15, 14, 12]}
df = pd.DataFrame(data)

# Create the hexbin plot
plt.hexbin(df['Price'], df['Mileage'], gridsize=10, cmap='inferno')
plt.title('Car Price vs. Mileage')
plt.xlabel('Price (USD)')
plt.ylabel('Mileage (MPG)')
cb = plt.colorbar()
cb.set_label('Count in Bin')
plt.show()
```

10. Create a 3D scatter plot of the provided dataset with horsepower on the x-axis, weight on the y-axis, and miles per gallon on the z-axis.

```python
import matplotlib.pyplot as plt
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D

# Generate sample data
data = {'Horsepower': [130, 165, 150, 250, 300],
        'Weight': [3000, 3500, 3200, 4200, 4800],
        'MPG': [25, 20, 22, 15, 12]}
df = pd.DataFrame(data)

# Create the 3D scatter plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['Horsepower'], df['Weight'], df['MPG'], c='r', marker='o')
ax.set_title('3D Scatter Plot of Car Attributes')
ax.set_xlabel('Horsepower')
ax.set_ylabel('Weight (lbs)')
ax.set_zlabel('Miles Per Gallon (MPG)')
plt.show()
```
